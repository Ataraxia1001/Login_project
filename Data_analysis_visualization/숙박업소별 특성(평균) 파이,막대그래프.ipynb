```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import font_manager , rc
mpl.rcParams['axes.unicode_minus']=False
plt.rc('font',family='Malgun Gothic')
```

```python
df = pd.read_excel('./업소종류별분석.xlsx')
```

```python
df[df['종류']=='프리미엄호텔']
```

```python
#파이 차트 그리기
plt.pie(df['평균가격'], labels=df['종류'], autopct='%1.1f%%' )

# 차트 제목 설정
plt.title('숙박업소종류별 평균가격비율')

# 차트 출력
plt.show()
```

```python
plt.bar(df['종류'], df['평균가격'] )

# X축 레이블 설정
plt.xlabel('숙박업소별 종류')

# Y축 레이블 설정
plt.ylabel('평균가격')

# 그래프 제목 설정
plt.title('숙박업소종류별 평균가격_막대그래프')

# 그래프 출력
plt.show()
```

```python
plt.bar(df['종류'], df['평균평점'])

# X축 레이블 설정
plt.xlabel('숙박업소별 종류')

# Y축 레이블 설정
plt.ylabel('평균평점')

# 그래프 제목 설정
plt.title('숙박업소종류별 평균평점_막대그래프')

# 그래프 출력
plt.show()
```

```python
plt.bar(df['종류'], df['평균도보거리(m)'])

# X축 레이블 설정
plt.xlabel('숙박업소별 종류')

# Y축 레이블 설정
plt.ylabel('평균도보거리(m)')

# 그래프 제목 설정
plt.title('숙박업소종류별 평균도보거리(m)_막대그래프')

# 그래프 출력
plt.show()
```
