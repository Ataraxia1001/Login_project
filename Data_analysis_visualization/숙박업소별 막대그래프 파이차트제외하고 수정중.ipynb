```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import font_manager , rc
mpl.rcParams['axes.unicode_minus']=False
plt.rc('font',family='Malgun Gothic')
```

```python
df = pd.read_excel('./업소종류별분석.xlsx')
```

```python
df[df['종류']=='프리미엄호텔']
```

```python
#파이 차트 그리기
plt.pie(df['평균가격'], labels=df['종류'], autopct='%1.1f%%' )

# 차트 제목 설정
plt.title('숙박업소종류별 평균가격비율')

# 차트 출력
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dataframe 'df' should be defined with the necessary data

# Set the figure size
plt.figure(figsize=(14, 6))  # Adjust the width and height as desired

df_sorted = df.sort_values('평균가격', ascending=False)

plt.bar(df_sorted['종류'], df_sorted['평균가격'])

# X-axis label
plt.xlabel('숙박업소별 종류', fontsize=15)

plt.xticks(rotation=45, ha='right', fontsize=15)

# Y-axis label
plt.ylabel('평균가격', fontsize=15)

plt.yticks(ha='right', fontsize=15)

# Graph title
plt.title('숙박업소종류별 평균가격_막대그래프', fontsize=15)

# Display the graph
plt.show()
```

```python
plt.figure(figsize=(14, 6))  # Adjust the width and height as desired

df_sorted = df.sort_values('평균평점', ascending=False)

plt.bar(df_sorted['종류'], df_sorted['평균평점'])

# X축 레이블 설정
plt.xlabel('숙박업소별 종류',fontsize=15)

plt.xticks(rotation=45, ha='right', fontsize=15)
# Y축 레이블 설정
plt.ylabel('평균평점',fontsize=15)

plt.yticks(ha='right', fontsize=15)
# 그래프 제목 설정
plt.title('숙박업소종류별 평균평점_막대그래프',  fontsize=15)

# 그래프 출력
plt.show()
```

```python
plt.figure(figsize=(14, 6))

df_sorted = df.sort_values('평균도보거리(m)', ascending=False)

plt.bar(df_sorted['종류'], df_sorted['평균도보거리(m)'])
# X축 레이블 설정
plt.xlabel('숙박업소별 종류', fontsize=15)

plt.xticks(rotation=45, ha='right', fontsize=15)

# Y축 레이블 설정
plt.ylabel('평균도보거리(m)',fontsize=15)

plt.yticks(ha='right', fontsize=15)

# 그래프 제목 설정
plt.title('숙박업소종류별 평균도보거리(m)_막대그래프', fontsize=15)

# 그래프 출력
plt.show()
```
